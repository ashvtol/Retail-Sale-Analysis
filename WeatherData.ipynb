```python
d2 = data
```

```python
d2.columns = ['Date', 'Temp', 'PCP24', 'sd']
```

```python
d2['Date'] = d2['Date'].astype('str')
```

```python
def convert(s):
    date = datetime.strptime(s,"%Y%m%d%H%M")
    return str(str(date.month )+ "/" + str(date.day) + "/" + str(date.year))
```

```python
d2['Date'] = d2['Date'].apply(lambda x: convert(x))
```

```python
d2['Temp'] = pd.to_numeric(d2['Temp'], errors = 'coerce')
```

```python
d2['PCP24'] = pd.to_numeric(d2['PCP24'], errors = 'coerce')
```

```python
d2['sd'] = pd.to_numeric(d2['sd'], errors = 'coerce')
```

```python
d2['Date'] = d2['Date'].astype('str')
```

```python
d3  = d2.groupby(['Date'])['Temp','PCP24','sd'].mean()
```

```python
d4 = pd.DataFrame(d3)
```
