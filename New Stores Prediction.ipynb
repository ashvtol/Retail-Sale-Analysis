```python
from uszipcode import SearchEngine
import statistics as stat
import warnings
import pandas as pd
warnings.simplefilter("ignore")
peepsPopForStore = []
IncomeForStore = []
storeData = []
filteredZip = []
fileterdNames = []
it = 0

for itr in StoreZip:
    search = SearchEngine(simple_zipcode=False)
    zipcode = search.by_zipcode(itr) 
    tempPop = 0
    temp = 0
    if(zipcode.state == 'NY'):
        for i in zipcode.population_by_age[2].values():
                if(temp == 1):
                    for val in i[4:]:
                        _,y = val.values()
                        tempPop += y
                temp += 1
        filteredZip.append(itr)
        IncomeForStore.append(zipcode.median_household_income)
        peepsPopForStore.append(tempPop)
        storeData.append((itr, zipcode.median_household_income, tempPop))
        fileterdNames.append(storeNames[it])
    it += 1
```

```python
populationThreshold = stat.median(peepsPopForStore)
incomeThreshold = stat.median(IncomeForStore)
```
